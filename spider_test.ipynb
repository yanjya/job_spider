```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#set page number find in loop
page_number = 2
url = f'https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=%E6%95%B8%E6%93%9A%E5%88%86%E6%9E%90&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=15&asc=0&page={page_number}&mode=l&jobsource=tab_cs_to_job&langFlag=0&langStatus=0&recommendJob=1&hotJob=1'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#find all the element 
elements = soup.find_all('article', class_='job-mode js-job-item')
job_link_elements =soup.find_all('a', class_='js-job-link')
```

```python
num = 1
for element in elements:
    # 這裡你可以對每個元素進行你想要的操作，例如打印元素的內容
    print(num)
    job_name = element['data-job-name']
    cust_name = element['data-cust-name']
    for link in job_link_elements:
        if link.text == job_name:
            job_link = link['href']
            break
    print('https:'+job_link)
    print(job_name)
    print(cust_name)

    num += 1
```
